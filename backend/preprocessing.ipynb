Mapping various sectors and sub-areas with no lat_long to their main areas' lat_long

```python
for sub_region in null_locs:
    for parent_region in lat_long_map.keys():
        if parent_region in sub_region and parent_region != sub_region:  # Check if parent region is a substring
            lat_long_map[sub_region] = lat_long_map[parent_region]
            print(f"Mapping {sub_region} to {parent_region}: {lat_long_map[sub_region]}")
            null_locs.remove(sub_region)
            break  # Stop checking once we've found a match
```

### Preprocessing Pipeline

```python
# Getting Latitude and Longitude 

from sklearn.base import BaseEstimator, TransformerMixin
import geopy
from geopy.geocoders import Nominatim


class LocationTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    
    def get_lat_long(self, location):
        geolocator = Nominatim(user_agent="geoapi")       
        print(f"fetching {location}")
        try:
            loc = geolocator.geocode(location + ", Mumbai")
            return loc.latitude, loc.longitude
        except:
            print(f"Error in fetching location {location}, searching MMR")
            try:
                loc = geolocator.geocode(location + ", Mumbai Metropolitan Region")
                return loc.latitude, loc.longitude
            except:
                print(f"Not even found in MMR") 
                try:
                    loc = geolocator.geocode(location + ", Navi Mumbai")
                    print(f"F")
                    return loc.latitude, loc.longitude
                except:
                    try:
                        loc = geolocator.geocode(location + ", District Thane")
                        print(f"Found in Thane")
                        return loc.latitude, loc.longitude
                    except:
                        print("Not found")  
                           
        return None, None
  

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        if isinstance(X, pd.Series):  
            X['lat_long'] = self.get_lat_long(X['Location'])
            X['Latitude'], X['Longitude'] = X['lat_long']
            X.drop('lat_long', inplace=True)
            X.drop('Location', inplace=True)

            return X
        else:
            X['lat_long'] = X['Location'].apply(self.get_lat_long)
            X.drop(columns=['Location'], inplace=True)
            X.dropna(subset=['lat_long'], inplace=True)
            X[['Latitude', 'Longitude']] = pd.DataFrame(X['lat_long'].tolist(), index=X.index)
            X.drop(columns=['lat_long', 'Location'], inplace=True)
            return X
```

```python
# Transforming Prices
class PriceTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    
    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        # Convert single row (Series) to DataFrame
        if isinstance(X, pd.Series):
            X = pd.DataFrame([X])
        
        X['Price )In Lakhs)'] = X['Price'] / 100000  # Assuming price is in rupees
        X.drop(columns=['Price','Unnamed: 0'], inplace=True)
        return X
```

```python
class LogTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.columns_to_transform = ['Area']  # Modify as necessary

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        # Convert single row (Series) to DataFrame
        if isinstance(X, pd.Series):
            X = pd.DataFrame([X])
        
        for col in self.columns_to_transform:
            if col in X.columns:
                X[col] = np.log1p(X[col])
        
        return X
```

```python
class FittedScaler(BaseEstimator, TransformerMixin):
    def __init__(self, scaler):
        self.scaler = scaler
    
    def fit(self, X, y=None):
        return self  # Return self, since it's already fitted
    
    def transform(self, X, y=None):
        return self.scaler.transform(X)
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline(steps=[
    ('location_transformer', LocationTransformer()),  # No fitting needed
    ('log_transformer',LogTransformer()),        # No fitting needed
    ('scaler', sc)                          # Already fitted scaler
])
```
